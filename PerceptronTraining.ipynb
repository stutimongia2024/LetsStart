#Approach 2 
I am super enthusiastic for learning ML so I will implement the model using ML
Let's do this :)

You will be needing `xTrain` , `yTrain`, `xTest` and `yTest` from the list `x` and `y` 

````python
print(f'This is x: {x} ' )

print(f'This is y: {y}' )
````

````python
xTrain=x [:-5] #training data
yTrain=y[:-5] #training labels 

xTest=x[-5:] #testing data
yTest=y[-5:] #testing labels 

print (f'''

xTrain : {xTrain}
yTrain : {yTrain}

xTest : {xTest}
yTest : {yTest}
 ''')
````

````python
import tensorflow as tf #framework
from tensorflow import keras #API / library 
````

````python
model = tf.keras.Sequential([
                             tf.keras.layers.Dense(units=1, input_shape=[1]) 
])
````

````python
model.compile(optimizer='adam', loss='mae') 
````

````python
numEpoch = 5000 #training it for 5000 times

model.fit(x=xTrain, y=yTrain, validation_data=(xTest, yTest), epochs=numEpoch)
````
